#ETAPA 3: Preparação dos dados



1.   Tratamento dos dados:

Após chegar à conclusão de que as colunas somente precisariam ser renomeadas, nenhum registro foi descartado e todo o dataset fornecido será ultilizado no projeto.



```python
#Renomeando as colunas para sanar o problema dectado nos dados:
dados = dados.rename(columns={'youtube': 'Sales','facebook':'Youtube','newspaper':'Facebook','sales':'Newspaper' })
dados.head()
```

2.   Construção dos dados:

Para um melhor entendimento e aproveitamento dos dados e suas correlações, serão adcionadas 2 novas colunas ao dataset:

**Investimento Total - Soma dos valores de Youtube, Facebook e Newspaper;**

**ROI(Return Over Investiment) - Que, utlizando a fórmula demonstrada logo abaixo, calcula o retorno total após a dedução do valor de um investimento.**

</br>
$$
ROI =\frac{sales-(youtube+facebook+newspaper)}{(youtube+facebook+newspaper)} X 100
$$
</br>


```python
#Inserindo a Coluna "Investimento" no dataframe principal:
dados_alt = dados
dados_alt['Investimento'] = dados_alt['Youtube'] + dados_alt['Facebook'] + dados_alt['Newspaper']
dados_alt.head()
```

```python
#Inserindo a Coluna "ROI" no dataframe principal:
dados_alt['ROI'] = round((dados_alt['Sales'] - dados_alt['Investimento']) / dados_alt['Investimento'],2) *100
dados_alt.head()
```

```python
#Atualizando o dataframe com os novos dados:
dados_total = pd.DataFrame({'Tipo': dados_alt.columns, 'Total':dados_alt.sum()})
dados_total.drop(['ROI'], inplace=True)
dados_total.reset_index(inplace=True)
dados_total = dados_total[['Tipo', 'Total']]
dados_total.sort_values(by='Total', ascending=False)
```

```python
#Inserindo os gráficos de linha de cada investimento:
multi = dados_alt[['Youtube', 'Facebook', 'Newspaper']]
plt.figure(figsize=(12,8))
plt.style.use('seaborn-v0_8-darkgrid')
palette = plt.get_cmap('Set1')
num = 0
for col in multi:
  num += 1
  plt.subplot(3, 1, num)
  for v in multi:
    plt.plot(multi.index, multi[v], marker='', color='red', linewidth=0.6, alpha=0.4)
  plt.plot(multi.index, multi[col], marker='', color=palette(num), linewidth=1.6, alpha=0.9, label=col)
  plt.title(col, loc='center', fontsize=8, fontweight=0, color=palette(num))
plt.show();
```

```python
#Preenchendo um gráfico de linhas com o total de Investimento X Vendas:
fig, ay = plt.subplots(figsize=(14, 6))
sns.lineplot(data=dados_alt, x=dados_alt.index, y='Investimento', color='red', label='Investimento')
sns.lineplot(data=dados_alt, x=dados_alt.index, y='Sales', color='green', label='Sales')
ay.fill_between(
    dados_alt.index, dados_alt['Investimento'], dados_alt['Sales'],
    where=(dados_alt['Investimento'] > dados_alt['Sales']),
    interpolate=True, color='red', alpha=0.25
    )
ay.fill_between(
    dados_alt.index, dados_alt['Investimento'], dados_alt['Sales'],
    where=(dados_alt['Investimento'] < dados_alt['Sales']),
    interpolate=True, color='green', alpha=0.25
    )
ay.legend()
plt.xlabel('Ocorrência')
plt.ylabel('Valor')
plt.title("Valor Total de Investimento com Vendas");
```

```python
#Verificando as correlações dos dados em tabela:
dados_alt.corr()
```

```python
#Verificando a correlação entre as colunas
sns.pairplot(dados_alt, x_vars=['Youtube', 'Facebook', 'Newspaper', 'Investimento'], y_vars=['Sales', 'ROI']);
```

```python
#Verificando as correlações com Mapa de Calor:
sns.heatmap(dados_alt.corr(),cmap='Reds');
```

#Entendendo os dados gerados:


*   **O impacto de investimento em youtube e facebook sobre as vendas é práticamente nulo. Ademais, quando comparado ao ROI, ambas as colunas mostram uma correlação decrescente, ou seja, quanto maior o investimento nelas, menor o retorno.**

*   **A melhor correlação encontrada é a do investimento em mídia impressa, que mostra uma crescente em relação às vendas e ao ROI.**

*   **Não importa tanto o quanto se investe em anúncios, mas, sim, onde.**



#ETAPA 4: Modelagem/Análise



1.   **Selecionando o Modelo:**
  
  O modelo selecionado para esse projeto foi o de Regressão Linear Simples do módulo SKLearn.

2.   **Treinando o Modelo:**

  O modelo será treinado separando o dataset tratado (dados_alt) em 4 variáveis:

  Uma X de treino e uma X de teste;

  Uma Y de treino e uma Y de teste.




```python
#Separando as variáveis X e y:
X = dados_alt[['Youtube', 'Facebook', 'Newspaper']]
y = dados_alt['Sales']
```

```python
#Dividindo as variáveis entre Treino e Teste:
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, random_state=22)
```

```python
#Treinando o modelo:
regL = LinearRegression()
regL.fit(X_train, y_train)
```



3.   **Teste do Modelo:**




```python
#Testando:
model_teste = regL.predict(X_test)
```

```python
#Verificando a acurácia:
acuracia = r2_score(y_test, model_teste)
print(f'A acurácia deste modelo é de {round(acuracia*100)}%.')
```
